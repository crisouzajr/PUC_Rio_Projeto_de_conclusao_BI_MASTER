### Verificar o como as escala das falhas ficam classificadas utilizando o modelo de árvore decisão.

#### 1 - Carregar o modelo treinado
#### 2 - Carregar os dados a serem inferidos
#### 3 - Fazer as inferências
#### 4 - FIM

```python
# Carregando modelo

import pickle
model = pickle.load(open('model_DT.sav', 'rb'))
```

```python
import pandas as pd

df = pd.read_excel('O_G_Equipment_Data.xlsx')
df.head(2)
```

```python
df = pd.get_dummies(df, columns = ['Preset_1', 'Preset_2'])
df.head(2)
```

```python
# Removendo as colunas para realizar a predição 
df = df.drop(columns={'Fail'})
df.head(2)
```

```python
# Inferência

inferences = model.predict(df)
df['Resultado'] = inferences
df.head(4)
```

```python
df_t = pd.DataFrame(df.Resultado.value_counts().sort_index())
df_t = df_t.reset_index()
df_t = df_t.rename({'index': 'Fail_Category'}, axis=1)
df_t
```
